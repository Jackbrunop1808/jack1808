# Label Encodinding

```python

from sklearn.preprocessing import LabelEncoder
LE=LabelEncoder()
```

```python
data['age_bracket'] = LE.fit_transform(data.age_bracket) 
data['race'] = LE.fit_transform(data.race)
data['sex'] = LE.fit_transform(data.sex)
data['census_msa'] = LE.fit_transform(data.census_msa)
data
```

# Train and Test

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_ind,y_dep,train_size=0.8,random_state=2)
```

# Logistic Regression


```python
from sklearn.linear_model import LogisticRegression
```

```python
Log=LogisticRegression()
```

```python
Log=LogisticRegression()
fit=Log.fit(x_train,y_train)
```

# Prediction

```python
Prediction=fit.predict(x_test)
Prediction
```

```python
df=pd.DataFrame({"Actual":y_test,"Predicted":Prediction})
df
```

```python
#Checking Accuracy Score
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
confusion_matrix(y_test,Prediction)
```

```python
accuracy_score(y_test,Prediction)
```

# Receiver Operating Characteristic (ROC) metric to evaluate output quality.

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
logit_roc_auc=roc_auc_score(y_test,fit.predict(x_test))
fpr,tpr,threshold=roc_curve(y_test,fit.predict_proba(x_test)[:,1])
plt.figure(figsize=(16, 16))
plt.plot(fpr,tpr,logit_roc_auc)
plt.legend(["H1N1"],loc=4)
plt.xlabel("fpr")
plt.ylabel("tpr")
```

```python
clf=LogisticRegression(class_weight="balanced")
clf.fit(x_train,y_train)
Threshold=0.8
```

```python
preds=np.where(clf.predict_proba(x_test)[:,1]>Threshold,1,0)
```

```python
pd.DataFrame(data=[accuracy_score(y_test,preds),roc_auc_score(y_test,preds)],
             index=["Accuracy","Roc_auc_score"])
```
