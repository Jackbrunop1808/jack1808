```python
## Checking Stationary 
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose as sd
from statsmodels.tsa.arima_model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
```

```python
result = sd(data["Sales"],model="multiplicative",freq=30)
fig=plt.figure()
fig=result.plot()
fig.set_size_inches(16,9)
```

```python
##Augmented Dickey_Fuller test #adfuller is used to check stationarity or non stationarity
from statsmodels.tsa.stattools import adfuller
```

```python
dftest=adfuller(data["Sales"],autolag="AIC")
```

```python
dftest
```

```python
dfout=pd.DataFrame(dftest[0:4],index=["teststats","pvalue","lag","Number of Observations used"])
```

```python
dfout
```

```python
#from pmdarima.arima import auto_arima
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
```

```python
plot_acf(data["Sales"]); # for MA q=0
plot_pacf(data["Sales"]); # for AR p=0
```

```python
model_arima=ARIMA(data.Sales,order=(0,0,0))
model_arima_fit=model_arima.fit()
print(model_arima_fit.aic)
```

```python
from pmdarima.arima import auto_arima
model =auto_arima(data.Sales, start_p=1, start_q=1,
                      test='adf',       # use adftest to find optimal 'd'
                      max_p=3, max_q=3, # maximum p and q
                      m=1,              # frequency of series
                      d=None,           # let model determine 'd'
                      seasonal=False,   # No Seasonality
                      start_P=0, 
                      D=0, 
                      trace=True,
                      error_action='ignore',  
                      suppress_warnings=True, 
                      stepwise=True)
```

```python
model.plot_diagnostics(figsize=(10,8))
plt.show()
```
