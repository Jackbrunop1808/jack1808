```python
#importing required libraries
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("bank-additional-full.csv",sep=';')
data
```

```python
# EDA
data.info()
```

```python
#checking null value
data.isnull().sum()
```

```python
#Label Encoder
from sklearn.preprocessing import LabelEncoder 
```

```python
Le=LabelEncoder()
```

```python
data['job']=Le.fit_transform(data.job)
data['marital']=Le.fit_transform(data.marital)
data['education']=Le.fit_transform(data.education)
data['default']=Le.fit_transform(data.default)
data['housing']=Le.fit_transform(data.housing)
data['loan']=Le.fit_transform(data.loan)
data['contact']=Le.fit_transform(data.contact)
data['month']=Le.fit_transform(data.month)
data['day_of_week']=Le.fit_transform(data.day_of_week)
data['poutcome']=Le.fit_transform(data.poutcome)
data['y']=Le.fit_transform(data.y)
```

```python
data.info()
```

```python
#correlation
corr=data.corr()
corr
```

```python
#heatmap
sb.heatmap(data,annot=True,cmap="Paired_r")
plt.show()
```

```python
#x_ind & y_dep
```

```python
x_ind=data.drop("y",axis=1)
x_ind
```

```python
y_dep=data.y
y_dep
```

```python
#Data Pre-processing
#to make a data normal
from sklearn.preprocessing import StandardScaler
```

```python
norm=StandardScaler()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_ind,y_dep,test_size=0.2,random_state=2)
```

```python
x_train=norm.fit_transform(x_train)
```

```python
x_test=norm.fit_transform(x_test)
data.shape
```

```python
from sklearn.svm import SVC
```

```python
model=SVC(kernel="linear")
```

```python
fit_model=model.fit(x_train,y_train)
```

```python
fit_model.n_support_
```

```python
y_pred=fit_model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
c=confusion_matrix(y_test,y_pred)
c
```

```python
acc=accuracy_score(y_test,y_pred)
acc
```

```python
#Hyperparameter
```

```python
kernel=["linear","rbf","poly","sigmoid"]
Accu =[]
for i in kernel:
    model=SVC(kernel=i,C=1.0)
    model_n=model.fit(x_train,y_train)
    #y_pred=model_n.predict(x_test)
    print("for kernal:",i)
    Accu.append(accuracy_score(y_test,model_n.predict(x_test)))
print(Accu)
```

Comparing with all the types of kernel "rbf" has best accuracy

```python
df=pd.DataFrame({"kernal":kernel,"AccuracyScore":Accu})
df
```

```python
plt.figure(figsize=(12,10))
sb.barplot(x="kernal",y="AccuracyScore",data=df)
```
